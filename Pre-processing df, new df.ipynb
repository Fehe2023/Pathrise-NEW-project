# If the missing value is a number (int, float), I can enter the mean of a column there

```python
program_duration_days_mean=df['program_duration_days'].mean()
```

```python
program_duration_days_mean
```

# Raplce missing value with mean
import numpy as np

```python
import numpy as np
```

```python
df ['program_duration_days']= df['program_duration_days'].replace (np.nan, program_duration_days_mean)
```

```python
df.info()
```

# another column float with missing value

```python
number_of_interviews_mean=df['number_of_interviews'].mean()
```

```python
number_of_interviews_mean
```

```python
df ['number_of_interviews']= df['number_of_interviews'].replace (np.nan, number_of_interviews_mean)
```

```python
df.info()
```

# To convert the object or Boolean to a Int value in the data frame,is 'get dummy' is used. That is, add 0 or 1 to each category

```python
df_dummy=pd.get_dummies(df[['highest_level_of_education','biggest_challenge_in_search',  'employment_status ', 'length_of_job_search','professional_experience','work_authorization_status', 'race','gender','pathrise_status','primary_track']])
df_dummy
```

```python
df= df.drop (['highest_level_of_education','biggest_challenge_in_search',  'employment_status ', 'length_of_job_search','professional_experience','work_authorization_status', 'race','gender','pathrise_status','primary_track'], axis=1)
```

```python
df= pd.concat([df, df_dummy], axis=1)
```
