# compare two object column

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data_Pathrise.csv')
```

```python
df.info()
```

```python
sns.boxplot(x= df['highest_level_of_education'], y=df['professional_experience'])
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df['highest_level_of_education'], y=df['professional_experience'])
plt.xlabel('Highest Level of Education')
plt.ylabel('Professional Experience')
plt.title('boxplot of Education vs Professional Experience')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df['primary_track'], y=df['professional_experience'])
plt.xlabel('primary_track')
plt.ylabel('Professional Experience')
plt.title('boxplot of primary track vs Professional Experience')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df['highest_level_of_education'], y=df['primary_track'])
plt.xlabel('Highest Level of Education')
plt.ylabel('primary_track')
plt.title('boxplot of Education vs primary_track')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x=df['primary_track'], y=df['length_of_job_search'])
plt.xlabel('primary_track')
plt.ylabel('length_of_job_search')
plt.title('Scatterplot of Job Search Length vs Primary Track')
plt.show()
```

# If we compare 2 columns with the target column

```python
sns.catplot(x='primary_track', y='length_of_job_search', hue='placed', data=df, aspect=5)
plt.xlabel('primary_track', fontsize=50)
plt.ylabel('length_of_job_search', fontsize=50)
plt.title('Scatterplot of Job Search Length vs Primary Track')
plt.show()
```

```python
sns.catplot(x='race', y='gender', hue='placed', data=df, aspect=5)
plt.xlabel('race', fontsize=50)
plt.ylabel('gender', fontsize=50)
plt.title('race vs gender')
plt.show()
```
